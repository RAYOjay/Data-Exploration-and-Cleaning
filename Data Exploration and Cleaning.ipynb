```python
# After seeing so much inconsistency in the data, it's advisable to drop all rows affected

data.dropna(axis=0, inplace=True)
```

```python
data.isnull().sum() #to check sum of duplicated values
```

```python
# To check for duplicated values

data.duplicated().sum()
```

```python
# to check the percentage of duplicated values

data.duplicated().sum() /len(data)*100
```

```python
#to drop duplicated data

data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
#to lock and remove dollar signfrom the first two values

data.loc[0, 'Price'] = data.loc[0, 'Price'].replace('$', '')
```

```python
data.loc[1, 'Price'] = data.loc[1, 'Price'].replace('$', '')
```

```python
data.head()
```

```python
# To filter all data with dollar sign

filtered_data = data[data['Price'].astype(str).str.contains('$')] 
```

```python
data.tail()
```

```python
#To convert Price from Object to Float data type

data['Price'] = data['Price'].astype('Float64')
```

```python
data.info()
```

```python
#to change invoiceDate to a dateand time datatype

data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
```

```python
data.info()
```

```python
data['Country'].value_counts()
```

```python
## to replace spme categories in country

data['Country'] = data['Country'].replace({'UK': 'United Kingdom', 'USA' : 'United State of America', 'RSA': 'South Africa'})
```

```python
data.head()
```

```python
# To extract all columns with object datatype

cat_columns = [col for col in data.select_dtypes('object')]
```

```python
cat_columns
```

```python
# data.select_dtypes('object'): This selects all columns in the DataFrame that have the data type 'object' (typically string columns).
#.columns: This retrieves the names of those columns as a list.
#[col for col in ...]: This iterates over each column name, assigning it to col.

for col in cat_columns:
    print(col)
    print(data[col].value_counts())
    print('------------------------')
```

```python
for col in cat_columns:
    data[col] = data[col].str.strip()
```
